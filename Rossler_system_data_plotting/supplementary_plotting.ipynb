```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.gridspec as gridspec
```

```python
# List of image filenames
image_I_ext = ['PLOTS_ROSSLER_DYNAMICS_PNG\I_ext_phase_plot_0.25.png',
               'PLOTS_ROSSLER_DYNAMICS_PNG\I_ext_phase_plot_0.5.png',
               'PLOTS_ROSSLER_DYNAMICS_PNG\I_ext_phase_plot_1.0.png',
               'PLOTS_ROSSLER_DYNAMICS_PNG\I_ext_phase_plot_1.5.png',
               'PLOTS_ROSSLER_DYNAMICS_PNG\I_ext_phase_plot_1.75.png'
               ]


image_k = ['PLOTS_ROSSLER_DYNAMICS_PNG\k_phase_plot_0.275.png',
           'PLOTS_ROSSLER_DYNAMICS_PNG\k_phase_plot_0.33.png',
           'PLOTS_ROSSLER_DYNAMICS_PNG\k_phase_plot_0.43.png',
           'PLOTS_ROSSLER_DYNAMICS_PNG\k_phase_plot_0.5.png',
           'PLOTS_ROSSLER_DYNAMICS_PNG\k_phase_plot_0.6.png',
           ]
```

```python

fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(2, 3)

labels = ['A', 'B', 'C', 'D', 'E']

# First row: 3 images
for i in range(3):
    ax = fig.add_subplot(gs[0, i])
    img = Image.open(image_I_ext[i])
    ax.imshow(img)
    ax.axis('off')
    ax.text(0.05, 0.95, labels[i], transform=ax.transAxes, fontsize=22, fontweight='bold', color='black', ha='center', va='top')

# Second row: 2 images, last subplot empty
for i in range(3, 5):
    ax = fig.add_subplot(gs[1, i - 3])
    img = Image.open(image_I_ext[i])
    ax.imshow(img)
    ax.axis('off')
    ax.text(0.05, 0.95, labels[i], transform=ax.transAxes, fontsize=22, fontweight='bold', color='black', ha='center', va='top')

# Last subplot (bottom right) left empty
ax_empty = fig.add_subplot(gs[1, 2])
ax_empty.axis('off')

plt.subplots_adjust(left=0, right=1, top=1, bottom=0, hspace=0, wspace=0)

# # Save the result as a single PDF
plt.savefig('Phase_plot_I_ext.pdf', format='pdf', bbox_inches='tight')
plt.savefig('Phase_plot_I_ext.svg', format='svg', bbox_inches='tight')
plt.savefig('Phase_plot_I_ext.png', format='png', bbox_inches='tight')
plt.savefig('Phase_plot_I_ext.eps', format='eps', bbox_inches='tight')
plt.show()
```
